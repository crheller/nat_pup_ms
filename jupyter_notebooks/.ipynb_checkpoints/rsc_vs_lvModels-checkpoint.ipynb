```python
# define correction method
corr_method = '_rm1' # rm1 means remove state variable with regression post hoc.
                     # rm2 means remove state variable by subtracting xforms model prediction from residuals

# load all noise corr for raw / pupil regressed data
rsc = ld.load_noise_correlation('rsc')
rsc_pr = ld.load_noise_correlation('rsc_pr'+corr_method)

# load all noise corr for various xforms models (lv regression - differen n LVs / hyperparameter constraints
rsc_lv1_a0 = ld.load_noise_correlation('rsc_pr_lvr'+corr_method, xforms_model=xlv1a0)
rsc_lv1_a01 = ld.load_noise_correlation('rsc_pr_lvr'+corr_method, xforms_model=xlv1a01)
rsc_lv1_a02 = ld.load_noise_correlation('rsc_pr_lvr'+corr_method, xforms_model=xlv1a02)
rsc_lv1_a03 = ld.load_noise_correlation('rsc_pr_lvr'+corr_method, xforms_model=xlv1a03)
rsc_lv2_a0 = ld.load_noise_correlation('rsc_pr_lvr'+corr_method, xforms_model=xlv2a0)
rsc_lv2_a01 = ld.load_noise_correlation('rsc_pr_lvr'+corr_method, xforms_model=xlv2a01)
rsc_lv2_a02 = ld.load_noise_correlation('rsc_pr_lvr'+corr_method, xforms_model=xlv2a02)
rsc_lv2_a03 = ld.load_noise_correlation('rsc_pr_lvr'+corr_method, xforms_model=xlv2a03)
rsc_lv3_a0 = ld.load_noise_correlation('rsc_pr_lvr'+corr_method, xforms_model=xlv3a0)
rsc_lv3_a01 = ld.load_noise_correlation('rsc_pr_lvr'+corr_method, xforms_model=xlv3a01)
rsc_lv3_a02 = ld.load_noise_correlation('rsc_pr_lvr'+corr_method, xforms_model=xlv3a02)
rsc_lv3_a03 = ld.load_noise_correlation('rsc_pr_lvr'+corr_method, xforms_model=xlv3a03)
```

```python
# overall noise correlations
f, ax = plt.subplots(1, 1, figsize=(10, 8))

ax.errorbar(-1, rsc['all'].mean(), yerr=rsc['all'].sem(),
                marker='.', markersize=10, color='k', label='raw', lw=2)
ax.errorbar(-1, rsc_pr['all'].mean(), yerr=rsc_pr['all'].sem(),
                marker='.', markersize=10, color='grey', label='pupil-corrected', lw=2)

ax.errorbar([0, 1, 2, 3], [rsc_lv1_a0['all'].mean(),
                          rsc_lv1_a01['all'].mean(),
                          rsc_lv1_a02['all'].mean(),
                          rsc_lv1_a03['all'].mean()], 
                          yerr=[rsc_lv1_a0['all'].sem(),
                                rsc_lv1_a01['all'].sem(),
                                rsc_lv1_a02['all'].sem(),
                                rsc_lv1_a03['all'].sem()], label='1-D LV', marker='.', markersize=10, lw=2)
ax.errorbar([0, 1, 2, 3], [rsc_lv2_a0['all'].mean(),
                          rsc_lv2_a01['all'].mean(),
                          rsc_lv2_a02['all'].mean(),
                          rsc_lv2_a03['all'].mean()], 
                          yerr=[rsc_lv2_a0['all'].sem(),
                                rsc_lv2_a01['all'].sem(),
                                rsc_lv2_a02['all'].sem(),
                                rsc_lv2_a03['all'].sem()], label='2-D LV', marker='.', markersize=10, lw=2)
ax.errorbar([0, 1, 2, 3], [rsc_lv3_a0['all'].mean(),
                          rsc_lv3_a01['all'].mean(),
                          rsc_lv3_a02['all'].mean(),
                          rsc_lv3_a03['all'].mean()], 
                          yerr=[rsc_lv3_a0['all'].sem(),
                                rsc_lv3_a01['all'].sem(),
                                rsc_lv3_a02['all'].sem(),
                                rsc_lv3_a03['all'].sem()], label='3-D LV', marker='.', markersize=10, lw=2)


ax.axhline(0, linestyle='--', color='k')
ax.legend()

ax.set_ylabel('Noise correlation')
ax.set_xlabel('Hyerparameter constraint')
ax.set_xticks([-1, 0, 1, 2, 3])
ax.set_xticklabels(['NA', 0.0, 0.1, 0.2, 0.3])

ax.set_title('Overall Noise Correlation')


# delta noise correlations
f, ax = plt.subplots(1, 1, figsize=(10, 8))

ax.errorbar(-1, (rsc['sp'] - rsc['bp']).mean(), yerr=(rsc['sp'] - rsc['bp']).sem(),
                marker='.', markersize=10, color='k', label='raw', lw=2)
ax.errorbar(-1, (rsc_pr['sp'] - rsc_pr['bp']).mean(), yerr=(rsc_pr['sp'] - rsc_pr['bp']).sem(),
                marker='.', markersize=10, color='grey', label='pupil-corrected', lw=2)

ax.errorbar([0, 1, 2, 3], [(rsc_lv1_a0['sp'] - rsc_lv1_a0['bp']).mean(),
                          (rsc_lv1_a01['sp'] - rsc_lv1_a01['bp']).mean(),
                          (rsc_lv1_a02['sp'] - rsc_lv1_a02['bp']).mean(),
                          (rsc_lv1_a03['sp'] - rsc_lv1_a03['bp']).mean()], 
                          yerr=[(rsc_lv1_a0['sp'] - rsc_lv1_a0['bp']).sem(),
                          (rsc_lv1_a01['sp'] - rsc_lv1_a01['bp']).sem(),
                          (rsc_lv1_a02['sp'] - rsc_lv1_a02['bp']).sem(),
                          (rsc_lv1_a03['sp'] - rsc_lv1_a03['bp']).sem()], 
                          label='1-D LV', marker='.', markersize=10, lw=2)
ax.errorbar([0, 1, 2, 3], [(rsc_lv2_a0['sp'] - rsc_lv2_a0['bp']).mean(),
                          (rsc_lv2_a01['sp'] - rsc_lv2_a01['bp']).mean(),
                          (rsc_lv2_a02['sp'] - rsc_lv2_a02['bp']).mean(),
                          (rsc_lv2_a03['sp'] - rsc_lv2_a03['bp']).mean()], 
                          yerr=[(rsc_lv2_a0['sp'] - rsc_lv2_a0['bp']).sem(),
                          (rsc_lv2_a01['sp'] - rsc_lv2_a01['bp']).sem(),
                          (rsc_lv2_a02['sp'] - rsc_lv2_a02['bp']).sem(),
                          (rsc_lv2_a03['sp'] - rsc_lv2_a03['bp']).sem()], 
                          label='2-D LV', marker='.', markersize=10, lw=2)
ax.errorbar([0, 1, 2, 3], [(rsc_lv3_a0['sp'] - rsc_lv3_a0['bp']).mean(),
                          (rsc_lv3_a01['sp'] - rsc_lv3_a01['bp']).mean(),
                          (rsc_lv3_a02['sp'] - rsc_lv3_a02['bp']).mean(),
                          (rsc_lv3_a03['sp'] - rsc_lv3_a03['bp']).mean()], 
                          yerr=[(rsc_lv3_a0['sp'] - rsc_lv3_a0['bp']).sem(),
                          (rsc_lv3_a01['sp'] - rsc_lv3_a01['bp']).sem(),
                          (rsc_lv3_a02['sp'] - rsc_lv3_a02['bp']).sem(),
                          (rsc_lv3_a03['sp'] - rsc_lv3_a03['bp']).sem()], 
                          label='3-D LV', marker='.', markersize=10, lw=2)

ax.axhline(0, linestyle='--', color='k')
ax.legend()

ax.set_ylabel('Small minus big')
ax.set_xlabel('Hyerparameter constraint')
ax.set_xticks([-1, 0, 1, 2, 3])
ax.set_xticklabels(['NA', 0.0, 0.1, 0.2, 0.3])

ax.set_title('Delta Noise Correlation')

plt.show()
```

## KEY TAKEAWAYS
